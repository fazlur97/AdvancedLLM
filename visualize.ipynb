```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "Model": ["Chronos", "TimeFS", "Uni2TS"],
    "All Stocks": [0.037, 0.019, -0.869 ],
    "Top 10%": [-0.009, -0.008, -0.045 ],
    "Bottom 10%": [0.093, -0.008, -1.193],
}

df = pd.DataFrame(data)

display(df)
```

```python

# Plotting
plt.figure(figsize=(10, 6))
bar_width = 0.25
x = range(len(df["Model"]))

plt.bar([i - bar_width for i in x], df["All Stocks"], width=bar_width, label="All Stocks")
plt.bar(x, df["Top 10%"], width=bar_width, label="Top 10%")
plt.bar([i + bar_width for i in x], df["Bottom 10%"], width=bar_width, label="Bottom 10%")

plt.xticks(x, df["Model"])
plt.ylabel("R² Out-of-Sample")
plt.title("Comparison of R² Out-of-Sample Across Models and Stock Segments")
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

models = ['OLS+H', 'OLS-3+H', 'PLS', 'PCR', 'ENet+H', 'GLM+H', 'RF', 'GBRT+H', 'NN1', 'NN2', 'NN3', 'NN4', 'NN5']
r2_all = [-34.86, 2.50, 2.93, 3.08, 1.78, 2.60, 3.28, 3.09, 2.64, 2.70, 3.40, 3.60, 2.79]
r2_top = [-54.86, 2.48, 1.84, 1.64, 1.90, 1.82, 4.80, 4.07, 2.77, 4.24, 4.73, 4.91, 4.86]
r2_bottom = [-19.22, 4.88, 5.36, 5.44, 3.94, 5.00, 5.08, 4.61, 4.37, 3.72, 5.17, 5.01, 3.58]

x = np.arange(len(models))
bar_width = 0.25

plt.figure(figsize=(12, 6))
plt.bar(x - bar_width, r2_all, width=bar_width, label='All', color='blue')
plt.bar(x, r2_top, width=bar_width, label='Top', color='teal')
plt.bar(x + bar_width, r2_bottom, width=bar_width, label='Bottom', color='gold')

plt.xticks(x, models, rotation=45)
plt.ylabel('$R^2_{OOS}$ (%)')
plt.title('Annual Out-of-Sample $R^2$ Performance by Model')
plt.legend()
plt.tight_layout()
plt.grid(True, axis='y')
plt.show()
```
